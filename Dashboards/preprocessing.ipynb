```python
import pandas as pd

df1 = pd.read_csv("/content/World Energy Consumption.csv")
df1
```

```python
df1 = df1[['iso_code',
           'country',
           'year',
           'biofuel_electricity',
           'hydro_electricity',
           'nuclear_electricity',
           'solar_electricity',
           'wind_electricity',
           'other_renewable_electricity',
           'coal_electricity',
           'gas_electricity',
           'oil_electricity',
           'biofuel_elec_per_capita',
           'hydro_elec_per_capita',
           'nuclear_elec_per_capita',
           'solar_elec_per_capita',
           'wind_elec_per_capita',
           'coal_elec_per_capita',
           'gas_elec_per_capita',
           'oil_elec_per_capita',
           'population'
          ]]
df1
```

```python
#loc for data verification

df1.loc[(df1['country'] == 'Brazil') & (df1['year'] == 2020)]
```

```python
df2 = df1.melt(id_vars=['iso_code', 'country', 'year', 'population'],
              var_name='fuel',
              value_name='production'
)

df2
```

```python
def split_type(row):
    if row['fuel'].find('per_capita') != -1:
        value = row['production']
    else:
        value = 0
    return value

df2['per_capita'] = df2.apply(split_type, axis=1)

df2
```

```python
#Clean percapita from production
def clean_production(row):
    if row['per_capita'] != 0:
        value = 0
    else:
        value = row['production']
    return value


df2['production'] = df2.apply(clean_production, axis=1)

df2
```

```python
#Clean name 'fuel'

df2['fuel'] = df2['fuel'].str.replace('_elec_per_capita', '')
df2['fuel'] = df2['fuel'].str.replace('_electricity', '')

df2['fuel'].unique()
```

```python
#Final window

df_final = df2.groupby(['iso_code', 'country', 'year', 'population', 'fuel']).sum().reset_index()

df_final.loc[(df_final['country'] == 'Brazil') & (df_final['year'] == 2018)]
```

```python
df_final.to_csv('rwenergy_dataset.csv',index=False)
```
